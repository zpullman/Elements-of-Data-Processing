```python
#import ....
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist, squareform
import matplotlib.pyplot as plt
import seaborn as sns
import json

#VAT algorithm provided for your use
def VAT(R):
    """

    VAT algorithm adapted from matlab version:
    http://www.ece.mtu.edu/~thavens/code/VAT.m

    Args:
        R (n*n double): Dissimilarity data input
        R (n*D double): vector input (R is converted to sq. Euclidean distance)
    Returns:
        RV (n*n double): VAT-reordered dissimilarity data
        C (n int): Connection indexes of MST in [0,n)
        I (n int): Reordered indexes of R, the input data in [0,n)
    """
        
    R = np.array(R)
    N, M = R.shape
    if N != M:
        R = squareform(pdist(R))
        
    J = list(range(0, N))
    
    y = np.max(R, axis=0)
    i = np.argmax(R, axis=0)
    j = np.argmax(y)
    y = np.max(y)


    I = i[j]
    del J[I]

    y = np.min(R[I,J], axis=0)
    j = np.argmin(R[I,J], axis=0)
    
    I = [I, J[j]]
    J = [e for e in J if e != J[j]]
    
    C = [1,1]
    for r in range(2, N-1):   
        y = np.min(R[I,:][:,J], axis=0)
        i = np.argmin(R[I,:][:,J], axis=0)
        j = np.argmin(y)        
        y = np.min(y)      
        I.extend([J[j]])
        J = [e for e in J if e != J[j]]
        C.extend([i[j]])
    
    y = np.min(R[I,:][:,J], axis=0)
    i = np.argmin(R[I,:][:,J], axis=0)
    
    I.extend(J)
    C.extend(i)
    
    RI = list(range(N))
    for idx, val in enumerate(I):
        RI[val] = idx

    RV = R[I,:][:,I]
    
    return RV.tolist(), C, I

```

```python
#Question 3.2 (a)
# https://towardsdatascience.com/pca-using-python-scikit-learn-e653f8989e60
special_trafficDf = pd.read_csv("special_traffic.csv")
special_traffic = special_trafficDf.sample(1000, random_state = 1)

from sklearn.preprocessing import StandardScaler
features = ['maximum_speed', 'speed_limit', 'average_speed', 'vehicle_class_1','vehicle_class_2','vehicle_class_3',
            'vehicle_class_4','vehicle_class_5','vehicle_class_6','vehicle_class_7','vehicle_class_8', 'vehicle_class_9',
            'vehicle_class_10','vehicle_class_11','vehicle_class_12','vehicle_class_13','bike', 'motorcycle']
# Separating out the features
x = special_traffic.loc[:, features].values
# Separating out the target
y = special_traffic.loc[:,['StrType']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
#principalComponents = pca.fit_transform(data_transformed[:1000,:]) 
# wanted to test if we got similar output to 3.3 (c) since i used different methods to standardise the data, I found that 
# using the minmaxscaler makes it much clearer clusters that depend on StrType
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDf, special_traffic[['StrType']]], axis = 1)

fig = plt.figure(1, figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set(title = '2 component PCA', xlabel='Principal Component 1', ylabel='Principal Component 2')
targets = ['Council Minor', 'Arterial']
colors = ['r', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['StrType'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
print("Variance explained by each PC",pca.explained_variance_ratio_)

#Question 3.2 (b)
# On most samples we see a greater variance in the Council Minor roads. PCA shows that Council Minor and Arterial roads are
# highly correlated.
# Seed 1: There appears to be 2 clusters, the dense linear cluster and the less dense cluster at the 'base' of the other cluster
#         It seems that StrType isn't splitting the data into clear clusters
#
# If we use the minmaxscaler instead of the standard the clusters are much clearer and apear to be based on StrType
# 

#Question 3.2 (c)
fig = plt.figure(2, figsize = (8,8))
RV, C, I = VAT(special_traffic[features])
x=sns.heatmap(RV,cmap='viridis',xticklabels=False,yticklabels=False)
x.set(title = 'All Features used', xlabel='Objects', ylabel='Objects')
plt.show()

fig = plt.figure(3, figsize = (8,8))
RV, C, I = VAT(principalDf)
x=sns.heatmap(RV,cmap='viridis',xticklabels=False,yticklabels=False)
x.set(title = 'Principal components', xlabel='Objects', ylabel='Objects')
plt.show()

# Question 3.2 (d)
# Using seed 1
# In the figure that has all features we can clearly see one giant cluster, this makes sense since its all based on road data.
# In the Principal components figure we again see a major cluster with 2 potential clusters within it.
```

```python
# Question 3.3 (a)
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

# use what we have from 3.2 to speed up working
# Need to scale to give equal importance of features (could use StandardScalar() as in 3.2 as well leads to similar solution)
data = special_trafficDf[features]
mms = MinMaxScaler()
mms.fit(data)
data_transformed = mms.transform(data)

# 
# use the inertia attribute to identify the sum of squared distances of samples to the nearest cluster centre.
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)

# Plot
fig = plt.figure(1,figsize = (8,8))
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()

# k = 2 looks to be the main point of inflection, this makes sense as its what we could see in 3.2
#Question 3.3 (b)
model = KMeans(n_clusters = 3, random_state = 1)
model.fit(data_transformed)
clust_labels = model.predict(data_transformed)
centres = model.cluster_centers_
kmeans = pd.DataFrame(clust_labels)
count = kmeans.groupby([0]).size()
fig = plt.figure(2)
count.plot.bar()
plt.xlabel('Clusters')
plt.ylabel('Number of points')
plt.title("eggy")

#Question 3.3 (c)
# Scatter plot with colour coded by StrType we then compare this with kmeans plot of clusters, if they match up we know
# that the data is grouped well by StrType.
# Here we use k = 3 as its on discussion forum

#Question 3.3 (d)
pca = PCA(n_components = 2).fit(data_transformed)
pca_2d = pca.transform(data_transformed)
plt.figure('Reference Plot',figsize = (8,8))
targets = ['Council Minor', 'Arterial']
palette=palette = ['red','blue']
colors=special_trafficDf.StrType.replace(to_replace=special_trafficDf.StrType.unique(),value=palette).tolist()
plt.scatter(pca_2d[:, 0], pca_2d[:, 1], c = colors)
plt.title('Reference Plot')
plt.legend(targets) #not sure why this doesnt work
kmeans = KMeans(n_clusters = 3, random_state = 1)
kmeans.fit(data_transformed)
plt.figure('K-means with 3 clusters',figsize = (8,8))
plt.scatter(pca_2d[:, 0], pca_2d[:, 1], c = kmeans.labels_)
plt.title('K-means with 3 clusters')
plt.show()

# Findings: We can see that kmeans does capture the clusters by StrType to some degree. However we have the wrong amount of 
#           clusters so this could impact the kmeans. Upon testing with n_clusters = 2 we still see that kmeans fails to 
#           capture the edge cases of the clusters.This is likely due to the elliptical shape of the cluster, this is something
#           that kmeans struggles with.
#           
```
